# Distance

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import numpy as np
```

```python
df = pd.read_csv('distance.csv')
df
```

```python
plt.scatter(df['Infrared'], df['Distance'])
plt.xlabel('Infrared Sensor')
plt.ylabel('Distance')
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(df['Infrared'].to_numpy().reshape(-1, 1))
X_train, X_test, y_train, y_test = train_test_split(X_poly, df['Distance'], test_size=0.2, shuffle=True)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
# Use metrics like mean squared error, R-squared, etc.
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
model.intercept_, model.coef_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, df['Distance'], test_size=0.0000000001, shuffle=True)
pred = model.predict(poly.transform(np.array(range(4000)).reshape(-1,1)))
plt.scatter(np.array(range(4000)), pred)
plt.scatter(X_train[:, 1], y_train)

plt.xlabel('Infrared Sensor')
plt.ylabel('Distance')
plt.legend()
```

# Force

```python
df = pd.read_csv('force.csv')
df
```

```python
plt.scatter(df['Sensor'], df['Force'])
plt.xlabel('Force Sensor')
plt.ylabel('Applied Force')
```

```python
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(df['Sensor'].to_numpy().reshape(-1, 1))
X_train, X_test, y_train, y_test = train_test_split(X_poly, df['Force'], test_size=0.2, shuffle=True)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
# Use metrics like mean squared error, R-squared, etc.
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
model.intercept_, model.coef_
```

```python
import numpy as np
X_train, X_test, y_train, y_test = train_test_split(X_poly, df['Force'], test_size=0.0000000001, shuffle=True)
pred = model.predict(poly.transform(np.array(range(3500)).reshape(-1,1)))
```

```python
plt.scatter(np.array(range(3500)), pred)
plt.scatter(X_train[:, 1], y_train)
plt.xlabel('Force Sensor')
plt.ylabel('Applied Force')
plt.legend()
```
